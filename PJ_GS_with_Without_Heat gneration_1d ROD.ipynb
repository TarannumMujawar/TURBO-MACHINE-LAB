```python
#Code 1 Point Jacobi Without Heat generation
import numpy as np
import matplotlib.pyplot as plt

# Define the problem parameters
L = 1.0  # Length of the rod
N = 10  # Number of grid points
dx = L / (N - 1)
T_left = 100.0  # Temperature at the left end
T_right = 25.0  # Temperature at the right end
tol = 1e-6  # Tolerance for convergence

# Initialize the temperature array
T = np.zeros(N)
T[0] = T_left
T[-1] = T_right

# Iterate using the point Jacobi method
iterations = 0
while True:
    T_old = T.copy()
    for i in range(1, N - 1):
        T[i] = (T[i - 1] + T[i + 1]) / 2.0
    iterations += 1
    if np.max(np.abs(T - T_old)) < tol:
        break

# Plot the temperature distribution
x = np.linspace(0, L, N)
plt.plot(x, T)
plt.xlabel("Position")
plt.ylabel("Temperature")
plt.title("Temperature Distribution")
plt.show()

# Print the number of iterations
print("Number of iterations:", iterations)
```

```python
#Code2 Point Jacobi with heat generation
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
L = 1.0  # Length of the rod
N = 10  # Number of grid points
dx = L / (N - 1)
T_left = 100.0  # Temperature at the left end
T_right = 25.0  # Temperature at the right end
q_dot = 100  # Internal heat generation rate (W/m³)
k = 1  # Thermal conductivity (W/mK)
tol = 1e-6  # Tolerance for convergence

# Initialize temperature array
T = np.zeros(N)
T[0] = T_left
T[-1] = T_right

# Iterate using the point Jacobi method
iterations = 0
while True:
    T_old = T.copy()
    for i in range(1, N - 1):
        T[i] = (T[i - 1] + T[i + 1] + dx**2 * q_dot / k) / 2.0
    iterations += 1
    if np.max(np.abs(T - T_old)) < tol:
        break

# Plot the temperature distribution
x = np.linspace(0, L, N)
plt.plot(x, T)
plt.xlabel("Position")
plt.ylabel("Temperature")
plt.title("Temperature Distribution with Heat Generation")
plt.show()

# Print the number of iterations
print("Number of iterations:", iterations)
# Create a 2D array for the heatmap
heatmap_data = np.array([T])  # Assuming a 1D temperature distribution

# Plot the heatmap
plt.imshow(heatmap_data, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Temperature')
plt.xlabel('Length')
plt.title('Temperature Distribution (Heatmap)')
plt.show()
```
